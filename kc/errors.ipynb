```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv", index_col="id")
```

```python
lever = 1800 # mm
data["abbe_x"] = np.arctan(data["dx"] / lever) * 180 / math.pi
data["abbe_z"] = np.arctan(data["dz"] / lever) * 180 / math.pi
```

```python
for i in data["abbe_x"]:
    print(f"{i:0.2f}")
```

```python
for i in data["abbe_z"]:
    print(f"{i:0.2f}")
```

```python
print(np.mean(data["dx"]), np.std(data["dx"]))
print(np.mean(data["dz"]), np.std(data["dz"]))
print(np.mean(data["abbe_x"]), np.std(data["abbe_x"]))
print(np.mean(data["abbe_z"]), np.std(data["abbe_z"]))
```

```python
import matplotlib

matplotlib.rcParams.update({'font.size': 22})

plt.figure(figsize=(8, 8), dpi=144)
plt.scatter(data["dx"], [np.mean(data["dz"])] * 10, alpha=0.3)
plt.scatter([np.mean(data["dx"])] * 10, data["dz"], alpha=0.3)
plt.gca().set_aspect("equal")
plt.xlim(-20, 0)
plt.ylim(-10, 10)
plt.xlabel("$\Delta x$")
plt.ylabel("$\Delta z$")
```
