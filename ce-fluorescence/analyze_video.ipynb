```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
vid = cv2.VideoCapture("/Users/mattfeng/Desktop/494A9868-ce-fluorescence.mov")
```

```python
fps = vid.get(cv2.CAP_PROP_FPS)
print("fps:", fps)
```

```python
saved_imgs = []
times = []
intensities = []

pos = (490, 960)
timestamps = set([528])

count = 0
success = True
vid.set(cv2.CAP_PROP_POS_FRAMES, 0)

while success:
    success, frame = vid.read()
    count += 1
    ts = count / fps
    
    intensities.append(frame[pos][1])
    times.append(ts)
```

```python
plt.figure(figsize=(9, 3), dpi=144)
plt.title("Pixel intensity over CE run")
plt.xlabel("time (s)")
plt.ylabel("pixel intensity (0-255)")
plt.xticks(np.arange(min(times), max(times)+1, 60))
plt.plot(times, intensities)
```
